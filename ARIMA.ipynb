# 1. ARIMA fitting and prediction for HOSE

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Read the data
hose_df_2022 = pd.read_csv('./data/hose_df_2022.csv')
hose_df_2023 = pd.read_csv('./data/hose_df_2023.csv')

# Convert 'date' column to datetime
hose_df_2022['date'] = pd.to_datetime(hose_df_2022['date'])
hose_df_2023['date'] = pd.to_datetime(hose_df_2023['date'])

# Sort the dataframes by 'ticker' and 'date'
hose_df_2022 = hose_df_2022.sort_values(by=['ticker', 'date'])
hose_df_2023 = hose_df_2023.sort_values(by=['ticker', 'date'])

# Step 1: Create an empty DataFrame with dates from hose_df_2023
dates_2023 = sorted(hose_df_2023['date'].unique())
hose_arima_pred = pd.DataFrame({'date': dates_2023})

# Add columns for each ticker
for ticker in hose_df_2022['ticker'].unique():
    hose_arima_pred[ticker] = None

# Set the 'date' column as the index
hose_arima_pred.set_index('date', inplace=True)

# Step 2: For loop for each ticker
for ticker in hose_df_2022['ticker'].unique():
    # Extract 'date' and 'close' price series for the current ticker from hose_df_2022
    ticker_data = hose_df_2022.loc[hose_df_2022['ticker'] == ticker, ['date', 'close']]
    
    # Set 'date' as the index
    ticker_data.set_index('date', inplace=True)
    
    # Fit an ARIMA(1,1,1) model
    model = ARIMA(ticker_data, order=(1, 1, 1))
    model_fit = model.fit()
    
    # Predict 249 days forward
    forecast = model_fit.forecast(steps=249)
    
    # Ensure the forecast index aligns with the prediction dates
    forecast_dates = hose_arima_pred.index[:249]  # Get the first 249 dates from hose_arima_pred
    
    # Fill the predicted values into the corresponding column of the DataFrame
    hose_arima_pred.loc[forecast_dates, ticker] = forecast.values

# Display the predictions DataFrame
print(hose_arima_pred)
hose_arima_pred.to_csv('hose_arima_pred.csv', index=True)



```
